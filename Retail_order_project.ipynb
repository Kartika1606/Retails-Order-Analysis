```python
#read data from the file 
import pandas as pd
df = pd.read_csv('orders.csv')
df.head(20)
```

```python
#handling null values
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#rename column names (making them lower case and replacing space with underscore 
#method1 = df.rename(columns = {'order ID' = 'order_id', 'City' = 'city'})

#method2 = convert column names to list of string objects
df.columns = df.columns.str.lower()

df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df.head(5)
```

```python
#convert oder date from object data type to datetime
#df.dtypes

df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
#derive new columns - discount , sale_price and profit
df['discount'] = df['list_price']*df['discount_percent']*0.01

df['sale_price'] = df['list_price'] - df['discount']

df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df.head(10)
```

```python
#drop columns - cost_price, list_price and discount_percent
df.drop(columns = ['cost_price','list_price','discount_percent'], inplace = True)
```
